```python
#Take a look at the raw data :
import pandas as pd
from pandas import read_csv
path = "../../data/raw/2010-2015_ili_activity_sub_climate.csv"

raw_data = read_csv(path)
raw_data.head()
```

```python
# display the variables
raw_data.info()
```

```python
#convert to datetime:
#https://stackoverflow.com/questions/38067704/how-to-change-the-datetime-format-in-pandas
raw_data['weekend'] = pd.to_datetime(raw_data.weekend)
```

```python
#rename column in pandas
#https://pandas.pydata.org/pandas-docs/version/0.21/generated/pandas.DataFrame.rename.html
raw_data.rename(index=str, columns={"weekend": "Date"}, inplace=True)
```

```python
# lower case
# https://stackoverflow.com/questions/19726029/python-make-pandas-dataframe-column-headers-all-lowercase
raw_data.columns = map(str.lower, raw_data.columns)
raw_data.info()
```
